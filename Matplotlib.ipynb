##Matplotlib
Matplotlib is a powerful Python library used for creating static, animated, and interactive visualizations. It offers a wide range of functionalities to create various types of plots and charts. Here's an overview of Matplotlib and its subtopics with simple examples:

###Matplotlib Overview
Matplotlib is divided into several submodules, each serving a different purpose:

pyplot: This submodule provides a MATLAB-like interface for creating and customizing plots. It's typically imported as plt.

Figure and Axes: The core objects in Matplotlib are the Figure and Axes.

A Figure object represents the entire figure or window where plots are drawn.
An Axes object represents a subplot (including the plot itself) which contains methods for plotting data.

####1. Line Plots

Line plots are used to visualize data points connected by straight line segments. They are typically used to show trends over a continuous interval or time series data.

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# Create a figure and axes
fig, ax = plt.subplots()

# Plotting
ax.plot(x, y, marker='o', linestyle='-', color='r', label='y = x^2')

# Customizing plot
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Line Plot Example')
ax.legend()

# Display plot
plt.show()
```

####2. Bar Charts
 Bar charts represent categorical data with rectangular bars where the length or height of each bar is proportional to the data it represents.

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [10, 30, 20, 15]

# Create a figure and axes
fig, ax = plt.subplots()

# Plotting
ax.bar(categories, values, color='c')

# Customizing plot
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Bar Chart Example')

# Display plot
plt.show()
```

####3. Scatter Plots
Scatter plots are used to visualize the relationship between two variables by displaying data points as dots on a 2D plane. Each dot represents one data point.

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
sizes = [20, 50, 80, 200]  # Marker sizes

# Create a figure and axes
fig, ax = plt.subplots()

# Plotting
ax.scatter(x, y, s=sizes, color='b', alpha=0.5)

# Customizing plot
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Scatter Plot Example')

# Display plot
plt.show()
```

####4. Histograms
 Histograms are used to represent the distribution of numerical data. It divides the data into bins and shows the frequency of each bin with bars.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data (random normal distribution)
data = np.random.randn(1000)

# Create a figure and axes
fig, ax = plt.subplots()

# Plotting
ax.hist(data, bins=30, color='green', edgecolor='black')

# Customizing plot
ax.set_xlabel('Values')
ax.set_ylabel('Frequency')
ax.set_title('Histogram Example')

# Display plot
plt.show()
```

####5. Pie Charts
Pie charts display the proportion of parts of a whole. Each slice of the pie represents a category's contribution to the total.

```python
import matplotlib.pyplot as plt

# Data
sizes = [25, 30, 15, 10]  # Sizes of each slice
labels = ['A', 'B', 'C', 'D']

# Create a figure and axes
fig, ax = plt.subplots()

# Plotting
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Customizing plot
ax.set_title('Pie Chart')

# Display plot
plt.show()
```

####6. Subplots
Subplots allow you to create multiple plots within the same figure. Each subplot is an independent plot that shares the same figure or canvas.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure and axes with 1 row and 2 columns
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Plotting on first subplot
ax1.plot(x, y1, color='c')
ax1.set_title('Sin Plot')

# Plotting on second subplot
ax2.plot(x, y2, color='g')
ax2.set_title('Cos Plot')

# Display plot
plt.show()
```

####Area Plot
An area plot (or filled line plot) is used to visualize quantitative data over a continuous interval, showing the relationship between variables and emphasizing the cumulative total.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(1, 11)
y1 = np.random.randint(1, 10, size=10)
y2 = np.random.randint(1, 10, size=10)

# Create a figure and axes
fig, ax = plt.subplots()

# Plotting
ax.fill_between(x, y1, color='skyblue', alpha=0.4, label='Y1 Area')
ax.fill_between(x, y2, color='orange', alpha=0.4, label='Y2 Area')

# Customizing plot
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Area Plot Example')
ax.legend()

# Display plot
plt.show()
```

####Box Plot
A box plot (or box-and-whisker plot) is a way of summarizing a set of data measured on an interval scale. It displays the distribution of data based on five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating some sample data
np.random.seed(42)
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(90, 20, 200)
data3 = np.random.normal(80, 15, 200)

data = [data1, data2, data3]

# Create a figure and axes
fig, ax = plt.subplots()

# Creating boxplot
ax.boxplot(data, patch_artist=True, showmeans=True)
ax.set_xticklabels(['Dataset 1', 'Dataset 2', 'Dataset 3'])
ax.set_title('Box Plot Example')
ax.set_ylabel('Values')

# Customizing fill colors and mean points
colors = ['lightblue', 'lightgreen', 'lightpink']
for patch, color in zip(ax.artists, colors):
    patch.set_facecolor(color)

# Display plot
plt.show()
```
